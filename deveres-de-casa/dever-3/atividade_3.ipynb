```python
def is_palindrome(arr, left=0, right=None):
    # Define o índice da direita na primeira chamada
    if right is None:
        right = len(arr) - 1

    # Caso base: cruzou ou encostou (é palíndromo)
    if left >= right:
        return True

    # Se os elementos forem diferentes, não é palíndromo
    if arr[left] != arr[right]:
        return False

    # Chamada recursiva avançando para o centro
    return is_palindrome(arr, left + 1, right - 1)


# Testes
array1 = [0, 1, 2, 3, 2, 1, 0]
array2 = ["a", "b", "b", "a"]
array3 = ["a", "b", "c", "b", "a"]
array4 = ["a", "b", "c", "f", "b", "a"]

print(is_palindrome(array1))  # True
print(is_palindrome(array2))  # True
print(is_palindrome(array3))  # True
print(is_palindrome(array4))  # False
```
