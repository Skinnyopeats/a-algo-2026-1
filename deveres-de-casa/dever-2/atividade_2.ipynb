```python
import time
import sys
sys.setrecursionlimit(2000)
```

```python
def factorial(n):
    """
    Calcula o fatorial de um número inteiro n utilizando recursão.
    """

    # Caso base da recursão
    if n == 0 or n == 1:
        return 1

    # Chamada recursiva
    return n * factorial(n - 1)
```

```python
def measure_execution(n):
    """
    Mede o tempo de execução do cálculo do fatorial.
    """

    start_time = time.perf_counter()

    result = factorial(n)

    end_time = time.perf_counter()

    execution_time = end_time - start_time

    print(f"n = {n}")
    print(f"Tempo de execução: {execution_time:.8f} segundos\n")

    return result
```

```python
test_values = [10, 100, 500, 1000]

for n in test_values:
    measure_execution(n)
```

## Análise de Complexidade

O algoritmo de cálculo do fatorial utiliza recursão.

A cada chamada da função `factorial(n)`, o algoritmo realiza:
- uma multiplicação
- uma chamada recursiva para `factorial(n-1)`

A relação de recorrência pode ser descrita como:

T(n) = T(n-1) + O(1)

Expandindo a recorrência:

T(n) = T(n-2) + 2c  
T(n) = T(n-3) + 3c  
...  

Portanto:

T(n) = O(n)

Assim, o tempo de execução cresce linearmente em relação ao tamanho da entrada `n`.

A complexidade de espaço também é O(n), pois cada chamada recursiva é armazenada na pilha de execução.
