```python
import random
import time
```

```python
def insertion_sort(arr):
    a = arr.copy()

    for i in range(1, len(a)):
        key = a[i]
        j = i - 1

        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1

        a[j + 1] = key

    return
```

```python
def medir_tempo(funcao, lista):
    inicio = time.time()
    funcao(lista)
    fim = time.time()
    return fim - inicio
```

```python
tamanhos = [1000, 5000, 10000, 20000, 50000]

for n in tamanhos:
    print(f"\nTamanho da lista: {n}")

    lista = [random.randint(0, 100000) for _ in range(n)]

    # Tempo Insertion Sort
    tempo_insertion = medir_tempo(insertion_sort, lista)
    print(f"Insertion Sort: {tempo_insertion:.6f} segundos")

    # Tempo sorted()
    tempo_sorted = medir_tempo(sorted, lista)
    print(f"sorted() (Timsort): {tempo_sorted:.6f} segundos")
```
